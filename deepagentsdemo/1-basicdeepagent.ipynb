```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ[""] = os.getenv("")
```

```python
from tavily import TavilyClient
from typing import Literal

tavily_client = TavilyClient(api_key = os.getenv())

def web_search(query:str, max_results: int = 5, topic: Literal["sports", "news", "Ai"] = "general", include_raw_content:bool=False):
    """Run a web Search"""
    return tavily_client.web_search(query=query, max_results=max_results, topic=topic, include_raw_content=include_raw_content)
    
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("groq: qwen/qwen3-32b")
model
```

```python
# Basic Agent
from langchain.agents import create_agent

simple_agent = create_agent(
    model=model,
    tools=[web_search]
)
simple_agent
```

```python
# Create a Deep Agent
# Prompt

# agent

from xml.parsers.expat import model

from deepagents import create_deep_agent

deepagent = create_deep_agent(
    models = model,
    tools = [web_search],
    system_prompt = "Act as a researcher"
)
deepagent
```

```python
result = deepagent.invoke({"message":[{"role":"user","content":"What is the latest news on AI?"}]})
result
```
